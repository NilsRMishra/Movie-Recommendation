```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
numerical_features = ["vote_average", "vote_count", "popularity", "runtime", "budget", "revenue"]
scaler = StandardScaler()
nndf = sampled_movies.copy()
nndf[numerical_features] = scaler.fit_transform(sampled_movies[numerical_features])
```

```python
nndf.shape
```

```python
def encode_text_feature(column, num_features=10):
    """Encodes categorical text features into fixed-size numerical lists."""
    nndf[column] = nndf[column].fillna("").apply(lambda x: x.split(",") if isinstance(x, str) else [])

    unique_values = set()
    for lst in nndf[column]:
        unique_values.update(lst)
    
    feature_map = {val: i+1 for i, val in enumerate(unique_values)}  # Avoid 0-indexing for better encoding

    encoded_list = []
    
    for row in nndf[column]:
        indices = [feature_map[val] for val in row if val in feature_map][:num_features]
        padded_indices = indices + [0] * (num_features - len(indices))  # Pad with zeros if less than num_features
        encoded_list.append(padded_indices)

    return encoded_list
```

```python
nndf["genres_encoded"] = encode_text_feature("genres", num_features=10)
nndf["keywords_encoded"] = encode_text_feature("keywords", num_features=10)

# Convert lists to NumPy arrays for PyTorch

X_genres = np.array(nndf["genres_encoded"])
X_keywords = np.array(nndf["keywords_encoded"])

print("Final Shape Check:", X_genres.shape, nndf.shape[0])  # Should match (num_movies, 10)

tfidf = TfidfVectorizer(max_features=100)
overview_embeddings = tfidf.fit_transform(nndf["overview"].fillna("")).toarray()
```

```python
X_numerical = torch.tensor(nndf[numerical_features].values, dtype=torch.float32)
X_genres = torch.tensor(np.array(nndf["genres_encoded"].tolist()), dtype=torch.float32)
X_keywords = torch.tensor(np.array(nndf["keywords_encoded"].tolist()), dtype=torch.float32)
X_overview = torch.tensor(overview_embeddings, dtype=torch.float32)
```

```python
# Combine into a Tensor dataset

dataset = TensorDataset(X_numerical, X_genres, X_keywords, X_overview)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
dataset[0]
```

```python
# Model Structure
class MovieEmbeddingModel(nn.Module):
    def __init__(self, input_size):
        super(MovieEmbeddingModel, self).__init__()
        self.fc1 = nn.Linear(input_size, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.output = nn.Linear(128, 25)  # 6-dimensional movie embeddings
        
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.relu(self.fc3(x))
        return self.output(x)  # Return final movie embedding

# Model Init
input_size = X_numerical.shape[1] + X_genres.shape[1] + X_keywords.shape[1] + X_overview.shape[1]
model = MovieEmbeddingModel(input_size)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Model Training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

num_epochs = 20
target_transform = nn.Linear(6, 25).to(device)  # Map 6D → 50D
```

```python
for batch in dataloader:
	temp_data = batch[0]
	break
```

```python
target_transform(temp_data).shape
```

```python
for epoch in range(num_epochs):
    for batch in dataloader:
        X_n, X_g, X_k, X_o = batch
        X_n, X_g, X_k, X_o = X_n.to(device), X_g.to(device), X_k.to(device), X_o.to(device)

        X_combined = torch.cat((X_n, X_g, X_k, X_o), dim=1)
        embeddings = model(X_combined)

        xformed_X_n = target_transform(X_n)
        # print(X_n.shape, embeddings.shape)
        # break
        # Loss computation (numerical features as truth-labels)
        loss = criterion(embeddings, xformed_X_n)

        # Backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
# ------------------ 4. Generate Movie Embeddings ------------------
with torch.no_grad():
    X_combined = torch.cat((X_numerical, X_genres, X_keywords, X_overview), dim=1).to(device)
    movie_embeddings = model(X_combined).cpu().numpy()
    
X_combined.shape
```

```python
# Recommend Similar Movies

indices = pd.Series(sampled_movies.index, index=nndf['title']).drop_duplicates()

def recommend_movies(title, top_n=5):
    """Finds the most similar movies based on cosine similarity."""
    if title not in indices:
        return "Movie not found in the sampled set."
    recoms = pd.DataFrame(columns=['Title', 'Ratings', 'Popularity', 'Genres'])
    movie_idx = nndf[nndf["title"] == title].index[0]
    movie_vector = movie_embeddings[movie_idx].reshape(1, -1)

    # Compute cosine similarity
    similarities = cosine_similarity(movie_vector, movie_embeddings)[0]
    similar_indices = similarities.argsort()[-top_n-1:-1][::-1]
    recoms = nndf[nndf['title'] == title][['title', 'genres']]
    recoms = pd.concat([recoms, nndf[['title', 'genres']].iloc[similar_indices]])
    return recoms
    # return nndf.iloc[similar_indices][["title", "genres", "vote_average"]]

# Example usage
print(recommend_movies("The Dark Knight", top_n=5))
```
